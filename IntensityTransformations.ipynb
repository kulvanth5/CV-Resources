```python
from google.colab import files
uploaded=files.upload()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread(list(uploaded.keys())[0], cv2.IMREAD_GRAYSCALE)

print(img.shape)
plt.imshow(img, cmap='gray')
plt.title("original Image")
plt.axis('off')
```

Negative Transformation
s = L-1-r

What we observe is that Dark background -> becomes bright and viceversa





```python
negative_img = 255- img

plt.figure(figsize=(10,9))
print(negative_img.shape)
plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(negative_img,cmap='gray')
plt.title("Negative Transformation")
plt.axis('off')

plt.show()
```

Log Transformation

s=clog(1+r)

Dark areas of the image become more visible

Bright areas dont overbright






```python
img_float=img.astype(np.float32)
c=255/np.log(1+np.max(img_float))
log_img=c*np.log(1+img_float)
log_img=np.array(log_img,dtype=np.uint8)
print(log_img.shape)

plt.figure(figsize=(10,10))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title("Original")
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(log_img, cmap='gray')
plt.title("Log Transformation")
plt.axis('off')
```

Gamma Correction (Power Law Transformation)

s=c*r*Gamma


Case-1

Gamma<1 (Brightening image)

```python
gamma=0.5
gamma_img1=np.power(img_float/255.0,gamma)
gamma_img1=np.uint8(gamma_img1*255)
print(gamma_img1.shape)

plt.figure(figsize=(10,10))

plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.title("Original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(gamma_img1,cmap='gray')
plt.title("Gamma Correction , G=0.5 ")
plt.axis('off')
plt.show()
```

Case-2

Gamma=1

```python
gamma=1
gamma_img2=np.power(img_float/255.0,gamma)
gamma_img2=np.uint8(gamma_img2*255)
print(gamma_img2.shape)



plt.figure(figsize=(10,10))

plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.title("Original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(gamma_img2,cmap='gray')
plt.title("Gamma Correction , G=1.0 ")
plt.axis('off')
plt.show()
```

Case-3:

Gamma>1 (Darkening image)

```python
gamma = 2.0   # gamma > 1
gamma_img3 = np.power(img_float / 255.0, gamma)
gamma_img3 = np.uint8(gamma_img3 * 255)
print(gamma_img3.shape)



plt.figure(figsize=(10,10))

plt.subplot(1,2,1)
plt.imshow(img,cmap='gray')
plt.title("Original")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(gamma_img3,cmap='gray')
plt.title("Gamma Correction , G=2.0 ")
plt.axis('off')
plt.show()
```
