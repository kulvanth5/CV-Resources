Original Image and its Histogram

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt


image = cv2.imread('/image.jpg')

gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(15, 10))

plt.subplot(2, 2, 1)
plt.title("1. Original Grayscale Image")
plt.imshow(gray_image, cmap='gray')
plt.axis('off')


plt.subplot(2, 2, 2)
plt.title("2. Original Histogram ")
plt.bar(np.arange(256), cv2.calcHist([gray_image], [0], None, [256], [0, 256]).flatten(),
width=0.8, color='steelblue')
plt.xlim([0, 256])

plt.tight_layout()
plt.show()
```

Equalized Image and its Histogram

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt


img = cv2.imread('/image.jpg', 0)


equ = cv2.equalizeHist(img)


res = np.hstack((img, equ))


plt.figure(figsize=(12, 6))


plt.subplot(2, 2, 1)
plt.title("Equalized Image")
plt.imshow(equ, cmap='gray')

plt.subplot(2, 2, 2)
plt.hist(equ.flatten(), 256, [0, 256], color='b')
plt.title("Equalized Histogram")

plt.tight_layout()
plt.show()
```
